## 根據範例，取出今天所有的新聞

```python
'''
Your Code
'''
from selenium import webdriver
from bs4 import BeautifulSoup

browser=webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")
```

```python
from datetime import datetime,timedelta

date_time=datetime.now()-timedelta(1)
date_time=date_time.strftime('%Y/%m/%d')
date_time
```

```python
'''
Your Code
'''
import time
while True:
    browser.execute_script("window.scrollTo(0, 100000);")
    time.sleep(2)
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    print(soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date").text)
    if date_time in soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date").text:
        break
```

## 取出今天下午三點到五點的新聞

```python
from datetime import datetime,timedelta
import time
date_timeH=datetime.now()
date_timeH=date_timeH.strftime('%Y/%m/%d')
date_timeH
```

```python
news=[]
for d in soup.find(class_="part_list_2").find_all('h3'):
    if datetime.strptime(d.find(class_="date").text, '%Y/%m/%d %H:%M') < date_time15H:
        break
    if datetime.strptime(d.find(class_="date").text, '%Y/%m/%d %H:%M') < date_time17H:
        news.append({
            '時間': d.find(class_="date").text,
            '標題': d.find_all('a')[-1].text
        })
import pandas as pd
pd.DataFrame(news)
```

## 根據範例，取出三天前下午三點到五點的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup

browser=webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")
```

```python
from selenium.webdriver.support.ui import Select
date_timeHT=datetime.now()-timedelta(3)
date_timeHM=date_timeHT.strftime('%m')
date_timeHD=date_timeHT.strftime('%d')

selM = Select(browser.find_element_by_id("selM"))
selM.select_by_value(date_timeHM)
selD = Select(browser.find_element_by_id("selD"))
selD.select_by_value(date_timeHD)

browser.find_element_by_id('button').click()
```

```python
import pandas as pd
date_timeHT=date_timeHT.strftime('%Y/%m/%d')
date_time15H=date_timeHT+' 15:00'
date_time17H=date_timeHT+' 17:00'
date_time15H=pd.to_datetime(date_time15H)
date_time17H=pd.to_datetime(date_time17H)
```

```python
date_time15H
```

```python
import time
while True:
    browser.execute_script("window.scrollTo(0, 100000);")
    time.sleep(1)
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    new_date = soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date").text
    print(new_date)
    if datetime.strptime(new_date, '%Y/%m/%d %H:%M') < date_time15H:
        break
```

```python
news=[]
for d in soup.find(class_="part_list_2").find_all('h3'):
    if datetime.strptime(d.find(class_="date").text, '%Y/%m/%d %H:%M') < date_time15H:
        break
    if datetime.strptime(d.find(class_="date").text, '%Y/%m/%d %H:%M') < date_time17H:
        news.append({
            '時間': d.find(class_="date").text,
            '標題': d.find_all('a')[-1].text
        })
import pandas as pd
pd.DataFrame(news)
```
