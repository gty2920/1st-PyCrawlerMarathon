## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

'''
Your Code
'''
import requests
import json
r=requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response=r.text
data=json.loads(response)
len(data)
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

'''
Your Code
'''
for d in data:
    print('標題:',d['title'])
    print('貼文時間:',d['createdAt'])
    print('留言人數:',d['commentCount'])
    print('按讚人數:',d['likeCount'])
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

'''
Your Code
'''
import requests
import json
fr=requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
fresponse=fr.text
fdata=json.loads(fresponse)
len(fdata)

FC=0
FL=0
TC=0
TL=0

for d in data:
    TC+=d['commentCount']
    TL+=d['likeCount']

for i in fdata:
    FC+=i['commentCount']
    FL+=i['likeCount']
    


print('熱門平均留言人數:',TC/len(data))
print('熱門平均按讚人數:',TL/len(data))
print('非熱門平均留言人數:',FC/len(fdata))
print('非熱門平均按讚人數:',FL/len(fdata))
```
